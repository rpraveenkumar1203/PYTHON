### Loading the Book 


```python
with open("miracle_in_the_andes.txt","r") as file:
    book = file.read()
```

```python
### commands to get count for specific words :

import re 

pattern = re.compile('Chapter [0-9]')
output = re.findall(pattern,book)
print(output)
```

```python

import re 

pattern = re.compile('Chapter [0-9]+')
#pattern = re.compile('Chapter [a-z]+')
output = re.findall(pattern,book)

print(output,len(output))
```

```python
# Need to get certain sentecne

pattern = re.compile('love')
output = re.findall(pattern,book)
print(output)
```

```python
pattern = re.compile('[a-zA-Z]* love [a-zA-Z]*')
output = re.findall(pattern,book)
print(output)
```

```python
pattern = re.compile('[a-zA-Z ,]* love [a-zA-Z]*')
output = re.findall(pattern,book)
print(output)
```

```python
pattern = re.compile('[^.]* love [^.]*')
output = re.findall(pattern,book)
print(output)
```

```python
### Most used words 

pattern  = re.compile("[a-zA-z]+")
findings = re.findall(pattern,book.lower())
findings1 = re.findall(pattern,book)
print(len(findings),len(findings1))
print(findings,findings1)
```

```python
repeating_words = {}

for word in findings:
    if word in repeating_words.keys():
        repeating_words[word] =  repeating_words[word] + 1
    else:
        repeating_words[word] = 1 

repeating_words
```
